```python
import pandas as pd
import glob
```

```python
files = glob.glob('data/processed_weather/varginha/*.csv')

dfs = []

for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

df_weather = pd.concat(dfs, ignore_index=True)
df_weather
```

```python
df_weather['datetime'] = pd.to_datetime(df_weather['datetime'])
df_weather = df_weather.sort_values('datetime')

df_weather = df_weather.drop_duplicates(subset='datetime')

df_weather.to_csv("data/weather.csv", index=False)
df_weather
```
